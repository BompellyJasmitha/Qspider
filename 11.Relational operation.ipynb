# Relational Operators : (<, >,<=, =>)

These are the operators which are used to find relation between Operands.

Types of relational operators:

        1. Greater than ( > )

        2. Less than( < )

        3. Greater than equals( >= )
        
        4. Less than equals (<= )

# --------------- A. GREATER THAN : ( > )----------------------------

It takes single value dataytypes & Multi value datatypes [ str, list, tuple ]

Here both Operands should be of same datatype

It gives result as TRUE if Operand-1 is greater than operand-2 , else give FALSE.

    Syntax: OP-1 > OP-2



```python
#Example: 1

a=10
b=20
s=a>b
print(s)
```

```python
#Example: 2

n1=10.5
n2=5
print(n1>n2)
```

```python
#Example: 3

c=10+5j
d=2
print(c>d)
```

```python
#Example: 4

st="python"
s=1
print(st>s)
```

```python
st='abc'
s='ABC'
print(st>s)
```

# Relational operations " > " in LIST

as done in strings it checks block by block if both are same then it checks for remaining blocks in order 

if 2 in 1 block & 5 in another block ,2>5 , result: FALSE  & vice versa





```python
# Example: 1

print([1,2,3]>[2,3])
print([1,2,3]>[1,1])
print(["raja"]>["rani"])
print(['ab','bc']>['ab','cd'])
print(['ab','bc']>['ab','bc'])

print(["raja"]>[1]) #Error
```
